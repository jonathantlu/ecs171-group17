```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")

print(df.shape)
df.head()
```

```python
fig, axes = plt.subplots(21, 1, figsize = (12, 12 * 21))

i = 0
for col in df.drop(columns = ['Diabetes_binary']):
    sns.histplot(df, x = col, hue = 'Diabetes_binary', ax = axes[i])
    i += 1

plt.show()
```
