```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")

print(df.shape)
df.head()
```

```python
diabetes = df.loc[df['Diabetes_binary'] == 1, :]
no_diabetes = df.loc[df['Diabetes_binary'] == 0, :]
for col in df.drop(columns = ['Diabetes_binary']):
    if df[col].min() == 0 and df[col].max() == 1:
        print(col, "is 1", diabetes[col].value_counts().to_dict()[1] / (diabetes[col].value_counts().to_dict()[1] + no_diabetes[col].value_counts().to_dict()[1]))
        print(col, "is 0", diabetes[col].value_counts().to_dict()[0] / (diabetes[col].value_counts().to_dict()[0] + no_diabetes[col].value_counts().to_dict()[0]))
```
